```python
import pandas as pd
import glob

# Define the directory containing the CSV files
data_dir = 'path/to/data'

# Use glob to find all CSV files in the directory
attack_file_paths = glob.glob(f'{data_dir}/*.csv')

# Use a generator expression to concatenate the files
data = pd.concat((pd.read_csv(f) for f in attack_file_paths), ignore_index=True)

# Add error handling
try:
    data = pd.concat((pd.read_csv(f) for f in attack_file_paths), ignore_index=True)
except FileNotFoundError as e:
    print(f"Error: {e}")
except pd.errors.EmptyDataError as e:
    print(f"Error: {e}")
    
    import os

# Create a dictionary to map filenames to attack types
attack_map = {
    'Data_of_Attack_Back_Normal.csv': 'Normal',
    'Data_of_Attack_Back.csv': 'Back',
    'Data_of_Attack_Back_BufferOverflow.csv': 'Buffer Overflow',
    # ... add more mappings for each file
}

# Create the "attack" column
data['attack'] = data.filename.apply(lambda x: attack_map[os.path.basename(x)])

from imblearn.over_sampling import RandomOverSampler
import numpy as np

# Create a RandomOverSampler instance with a robust random state
random_state = np.random.RandomState(42)
resampler = RandomOverSampler(random_state=random_state)

# Resample the data to address class imbalance
data_resampled, _ = resampler.fit_resample(data.drop('attack', axis=1), data['attack'])

# Add a comment to explain the purpose of the code
# Resample the data to address class imbalance, which can improve the performance of machine learning models

# Create a binary indicator column for attacks (non-Normal)
data_resampled['is_attack'] = data_resampled['attack'] != 'Normal'

# Standardize attack type names (e.g., convert to title case)
data_resampled['attack'] = data_resampled['attack'].map({
    'Normal': 'normal',
    'Back': 'Back',
    'BufferOverflow': 'Buffer Overflow',
    'FTPWrite': 'FTP Write',
    'GuessPassword': 'Guess Password',
    'Neptune': 'Neptune',
    'NMap': 'N-Map',
    'PortSweep': 'Port Sweep',
    'RootKit': 'Root Kit',
    'Satan': 'Satan',
    'Smurf': 'Smurf'
})
```
