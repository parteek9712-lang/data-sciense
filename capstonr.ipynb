```python
import pandas as pd

# Load the data from the 10 tables
data_back_normal = pd.read_csv('Data_of_Attack_Back_Normal.csv')
data_back = pd.read_csv('Data_of_Attack_Back.csv')
data_back_bufferoverflow = pd.read_csv('Data_of_Attack_Back_BufferOverflow.csv')
data_back_ftpwrite = pd.read_csv('Data_of_Attack_Back_FTPWrite.csv')
data_back_guesspassword = pd.read_csv('Data_of_Attack_Back_GuessPassword.csv')
data_back_neptune = pd.read_csv('Data_of_Attack_Back_Neptune.csv')
data_back_nmap = pd.read_csv('Data_of_Attack_Back_NMap.csv')
data_back_portsweep = pd.read_csv('Data_of_Attack_Back_PortSweep.csv')
data_back_rootkit = pd.read_csv('Data_of_Attack_Back_RootKit.csv')
data_back_satan = pd.read_csv('Data_of_Attack_Back_Satan.csv')
data_back_smurf = pd.read_csv('Data_of_Attack_Back_Smurf.csv')

# Append the data
data = pd.concat([data_back_normal, data_back, data_back_bufferoverflow, data_back_ftpwrite, 
                 data_back_guesspassword, data_back_neptune, data_back_nmap, data_back_portsweep, 
                 data_back_rootkit, data_back_satan, data_back_smurf], ignore_index=True)

# Create a new column "attack"
data['attack'] = data.apply(lambda row: row.name.split('_')[3], axis=1)

from imblearn.over_sampling import RandomOverSampler

# Perform resampling
ros = RandomOverSampler(random_state=42)
data_resampled, _ = ros.fit_resample(data, data['attack'])

# Create a new column "attack" for binomial classification
data_resampled['attack'] = data_resampled['attack'].apply(lambda x: 1 if x != 'Normal' else 0)

# Create a new column "attack" for multinomial classification
attack_map = {'Normal': 0, 'Back': 1, 'Buffer Overflow': 2, 'FTP Write': 3, 'Guess Password': 4, 
              'Neptune': 5, 'N-Map': 6, 'Port Sweep': 7, 'Root Kit': 8, 'Satan': 9, 'Smurf': 10}
data_resampled['attack'] = data_resampled['attack'].map(attack_map)
```
